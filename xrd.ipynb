```python
#import
from pymatgen.ext.matproj import MPRester 
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
#Téléchargement du fichier CIF & création de la structure
with MPRester("VGzmabHyhs7K1L84") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-7017")
    w = CifWriter(structure,0.1)
    w.write_file('mystructure.cif')
```

```python
# Création de l'object XRDCalculator pour un rayonnement X incidant Cu Ka de longueur d’onde λ=1.54060 Å 
xrd = XRDCalculator(1.54060)  

# Modèle de diffraction

model = xrd.get_pattern(structure)
```

```python
#Premier pic de diffraction

# Indices hkl associés au premier pic de diffraction
hkl_1 = model.hkls[0]

# Angle du premier pic de diffraction
angle_1 = model.x[0]

# Intensité du premier pic de diffraction
I_1 = model.y[0]

print("Premier pic de diffraction :\n")

print("Indices hkl correspondant : ", hkl_1)
print("Angle : 2*theta =", angle_1,"°")
print("Intensité :", I_1)
```

```python
#Deuxième pic de diffraction 

# Indices hkl associés au premier pic de diffraction
hkl_2 = model.hkls[1]

# Angle du premier pic de diffraction
angle_2 = model.x[1]

# Intensité du premier pic de diffraction
I_2 = model.y[1]

print("Deuxième pic de diffraction :\n")

print("Indices hkl correspondant : ", hkl_2)
print("Angle : 2*theta =", angle_2,"°")
print("Intensité :", I_2)
```

```python
#Troisième pic de diffraction

# Indices hkl associés au premier pic de diffraction
hkl_3 = model.hkls[2]

# Angle du premier pic de diffraction
angle_3 = model.x[2]

# Intensité du premier pic de diffraction
I_3 = model.y[2]

print("Troisième pic de diffraction :\n")

print("Indices hkl correspondant : ", hkl_3)
print("Angle : 2*theta =", angle_3,"°")
print("Intensité :", I_3)
```

```python
# Diagramme de diffraction

print("Le graphe ci_dessous réprésente les 3 premiers pics de diffractions pour un rayonnement X incident Cu Ka de longueur d'onde λ=1.54060 Å. Ce diagramme permet de vérifier les valeurs trouvées ci_dessus.")
xrd.get_plot(structure, two_theta_range=(15, 34), annotate_peaks=True, ax=None, with_labels=True, fontsize=10)
```
